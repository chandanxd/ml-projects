# Train-Test Split

Always evaluate your model on a test set that hasn't been used in the training process at all. So we'll split off a test set here.

_(Note: This might be a little less true if you use cross-validation, but is still considered a good practice.)_


```python
from sklearn.model_selection import train_test_split
```

```python
df_train, df_test = train_test_split(df, test_size=0.25)
```

```python
df_train.shape
```

```python
df_test.shape
```

```python
df_train.head()
```

# Prepare data for modeling

This involves splitting the back out into plain NumPy arrays.


```python
x_train = df_train.drop(columns=["target", "target_name"]).values
y_train = df_train["target"].values
```

```python
y_train
```

# Modeling - What is our baseline?

What is the simplest model we can think of?

In this case, if our model is just randomly guessing the species of the flower, or guessing a single species for every data point, we would expect the model to have an accuracy of 0.33 or 33%, since we have three different classes that are evenly balanced (3 classes  50 samples).

So our models should at least beat 0.33 or 33% accuracy.


# Modeling - A simple manual model

Let's manually look at our data & decide some cutoff points for our classification.


```python
def single_feature_prediction(petal_length):
    """Predicts the Iris flower species given the petal length"""
    if petal_length < 2.5:
        return 0
    elif petal_length < 4.8:
        return 1
    else:
        return 2
```

```python
x_train[:, 2]
```

```python
manual_y_predictions = np.array(
    [single_feature_prediction(val) for val in x_train[:, 2]]
)
```

```python
manual_model_accuracy = np.mean(manual_y_predictions == y_train)
```

```python
print(f"Accuracy for manual model: {manual_model_accuracy * 100:.2f}%")
```

# Modeling - Logistic Regression


```python
from sklearn.linear_model import LogisticRegression
```

### Using a validation set to evaluate our model


```python
model = LogisticRegression()
```

```python
# xt stands for "x_train" and xv stands for "x_validation"
xt, xv, yt, yv = train_test_split(x_train, y_train, test_size=0.25)
```

```python
xt.shape
```

```python
xv.shape
```

```python
model.fit(xt, yt)
```

```python
y_pred = model.predict(xv)
```

```python
np.mean(y_pred == yv)
```

```python
model.score(xv, yv)
```

### Using cross-validation to evaluate our model


```python
from sklearn.model_selection import cross_val_predict, cross_val_score
```

```python
model = LogisticRegression(max_iter=200)
```

```python
accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
```

```python
np.mean(accuracies)
```

# Where are misclassifying points?


```python
y_pred = cross_val_predict(model, x_train, y_train, cv=5)
```

```python
predicted_correctly_mask = y_pred == y_train
```

```python
not_predicted_correctly = ~predicted_correctly_mask
```

```python
x_train[not_predicted_correctly]
```

```python
df_predictions = df_train.copy()
```

```python
df_predictions["correct_prediction"] = predicted_correctly_mask
```

```python
df_predictions["prediction"] = y_pred
```

```python
df_predictions["prediction_label"] = df_predictions["prediction"].map(
    {0: "setosa", 1: "versicolor", 2: "virginica"}
)
```

```python
df_predictions.head()
```

```python
sns.scatterplot(
    x="petal length (cm)",
    y="petal width (cm)",
    hue="prediction_label",
    data=df_predictions,
    palette=species_colors,
)
```

```python
sns.scatterplot(
    x="petal length (cm)",
    y="petal width (cm)",
    hue="target_name",
    data=df_predictions,
    palette=species_colors,
)
```

```python
def plot_incorrect_prediction(df_predictions, x_axis_feature, y_axis_feature):
    fig, axs = plt.subplots(2, 2, figsize=(10, 10))
    axs = axs.flatten()
    sns.scatterplot(
        x=x_axis_feature,
        y=y_axis_feature,
        hue="prediction_label",
        data=df_predictions,
        ax=axs[0],
        palette=species_colors,
    )
    sns.scatterplot(
        x=x_axis_feature,
        y=y_axis_feature,
        hue="target_name",
        data=df_predictions,
        ax=axs[1],
        palette=species_colors,
    )
    sns.scatterplot(
        x=x_axis_feature,
        y=y_axis_feature,
        hue="correct_prediction",
        data=df_predictions,
        ax=axs[2],
    )
    axs[3].set_visible(False)

    plt.show()
```

```python
plot_incorrect_prediction(df_predictions, "petal length (cm)", "petal width (cm)")
```

# Model Tuning

What is model tuning?

Model tuning is trying to determine the parameters of your model (also known as "hyperparameters) that maximize the model performance.


```python
for reg_param in (1, 1.3, 1.8, 2, 2.3, 2.8, 3):
    model = LogisticRegression(max_iter=200, C=reg_param)
    accuracies = cross_val_score(model, x_train, y_train, cv=5, scoring="accuracy")
    print(f"Accuracy: {np.mean(accuracies) * 100}%")
```

# Final Model


```python
model = LogisticRegression(max_iter=200, C=2)
```

# How well does our model do on the test set?


```python
x_test = df_test.drop(columns=["target", "target_name"]).values
y_test = df_test["target"].values
```

```python
x_test.shape
```

```python
y_test
```

### Train our final model using our full training dataset


```python
model.fit(x_train, y_train)
```

```python
y_test_pred = model.predict(x_test)
```

```python
test_set_correctly_classified = y_test_pred == y_test
test_set_accuracy = np.mean(test_set_correctly_classified)
```

```python
print(f"Test set accuracy: {test_set_accuracy * 100:.2f}%")
```

```python
df_predictions_test = df_test.copy()
df_predictions_test["correct_prediction"] = test_set_correctly_classified
df_predictions_test["prediction"] = y_test_pred
df_predictions_test["prediction_label"] = df_predictions_test["prediction"].map(
    {0: "setosa", 1: "versicolor", 2: "virginica"},
)
```

```python
plot_incorrect_prediction(
    df_predictions_test,
    x_axis_feature="petal length (cm)",
    y_axis_feature="petal width (cm)",
)
```
